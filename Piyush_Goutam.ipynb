```python
# Simple Corrosion Risk Assessment Program

humidity = float(input("Enter Relative Humidity (%): "))
temperature = float(input("Enter Temperature (°C): "))
pH = float(input("Enter pH Value: "))
salt = input("Is Salt Present? (Yes/No): ").strip().lower()

if humidity > 70 and temperature > 30 and pH < 6 and salt == "yes":
    risk = "HIGH"
elif humidity > 50 and pH < 7:
    risk = "MODERATE"
else:
    risk = "LOW"

print("\nCorrosion Risk:", risk)
```
