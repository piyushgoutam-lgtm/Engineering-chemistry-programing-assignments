```python
plastics = {
    1: ["PET", "Water bottles, Soft drink bottles", "Highly Recyclable", "Safe and widely recycled"],
    2: ["HDPE", "Milk bottles, Detergent bottles", "Highly Recyclable", "Environment friendly"],
    3: ["PVC", "Pipes, Wire insulation", "Difficult to recycle", "Can release toxic chemicals"],
    4: ["LDPE", "Plastic bags, Food wraps", "Limited recycling", "Produces less pollution"],
    5: ["PP", "Food containers, Bottle caps", "Recyclable", "Safe for food storage"],
    6: ["PS", "Disposable cups, Plates", "Rarely recycled", "Not eco-friendly"],
    7: ["Others", "Mixed plastics", "Not easily recyclable", "Avoid if possible"]
}

code = int(input("Enter Recycling Code (1-7): "))

if code in plastics:
    print("\nPolymer Name:", plastics[code][0])
    print("Common Products:", plastics[code][1])
    print("Recyclability:", plastics[code][2])
    print("Environmental Remarks:", plastics[code][3])
else:
    print("Invalid Recycling Code!")
```
